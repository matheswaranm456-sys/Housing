### Outliers

```python
#fix outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['sqft_living'])
plt.title("Boxplot")
plt.show()
```

```python
Q1 = df['sqft_living'].quantile(0.25)
Q3 = df['sqft_living'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['sqft_living'] = np.where(df['sqft_living'] < lower_bound, lower_bound, df['sqft_living'])
df['sqft_living'] = np.where(df['sqft_living'] > upper_bound, upper_bound, df['sqft_living'])
print(df['sqft_living'])
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['sqft_living'])
plt.title("Boxplot")
plt.show()
```

```python
#price

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price'])
plt.title("Boxplot")
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['price'] = np.where(df['price'] < lower_bound, lower_bound, df['price'])
df['price'] = np.where(df['price'] > upper_bound, upper_bound, df['price'])
print(df['price'])
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price'])
plt.title("Boxplot")
plt.show()
```

```python
#Bedrooms:

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bedrooms'])
plt.title("Boxplot")
plt.show()
```

```python
# Fix the outliers

Q1 = df['bedrooms'].quantile(0.25)
Q3 = df['bedrooms'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['bedrooms'] = np.where(df['bedrooms'] < lower_bound, lower_bound, df['bedrooms'])
df['bedrooms'] = np.where(df['bedrooms'] > upper_bound, upper_bound, df['bedrooms'])
print(df['bedrooms'])

```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['bedrooms'])
plt.title("Boxplot")
plt.show()
```

```python
#Bathrooms:

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bathrooms'])
plt.title("Boxplot")
plt.show()
```

```python
# Fix the outliers

Q1 = df['bathrooms'].quantile(0.25)
Q3 = df['bathrooms'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mean_value = df['bathrooms'].mean()

df['bathrooms'] = np.where((df['bathrooms'] < lower_bound) | (df['bathrooms'] > upper_bound),mean_value,df['bathrooms'])
print(df['bathrooms'])
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['bathrooms'])  
plt.title("Boxplot")  
plt.show()
```

## OVERVIEW:
Missing Values:

The dataset initially had missing values in sqft_living, sqft_lot, yr_built, and city. These were addressed by dropping the city column (as it had many unique values and could be handled differently if needed for specific analysis) and imputing the median for the numerical columns (sqft_living, sqft_lot, yr_built).

#### Outliers:

Outliers were identified and treated in sqft_lot, bedrooms, sqft_living, and price columns using the IQR method, where extreme values were capped at the upper and lower bounds. This ensures that extreme values do not disproportionately influence analysis and models.

#### Data Types:

The date column was converted to datetime objects, which is crucial for time-series analysis and feature engineering.

#### Factors Influencing Housing Prices (Insights from EDA):

#### Square Footage (sqft_living):

There is a strong positive correlation between sqft_living and price (correlation coefficient of approximately 0.65). This indicates that larger living spaces tend to command higher prices, which is a fundamental principle in real estate.

#### Number of Bedrooms and Bathrooms:

While not as strongly correlated as sqft_living, bedrooms and bathrooms also show a positive relationship with price. The pairplot revealed clear trends where houses with more bedrooms and bathrooms generally have higher prices. The violin plot of bedrooms vs. bathrooms suggests a reasonable distribution, with most houses having 3-4 bedrooms and 1.5-2.5 bathrooms.

#### Floors and Condition:

The Chi-Square test between floors and condition showed a significant relationship (p-value < 0.05). This suggests that the number of floors might be associated with the general condition of the house, which in turn could influence pricing, although the direct relationship with price was not explicitly tested here.

#### Price Distribution:

The initial price distribution was right-skewed with many outliers (high-priced properties). After outlier treatment, the distribution became more normalized, which is beneficial for many statistical models.

#### Time Trends:

The line plot of price trends over time (date) did not show a very distinct upward or downward trend within the observed period. This might indicate stability or cyclical patterns that are not immediately apparent over this specific short timeframe.

#### Feature Engineering:

New features like age_of_property (derived from sale_year and yr_built) and price_per_sqft were created. price_per_sqft is a particularly insightful feature, providing a standardized measure of value that can be directly compared across properties of different sizes.

#### Conclusion:

The analysis confirms that common real estate drivers such as sqft_living, bedrooms, bathrooms, and premium features like waterfront access are significant factors influencing housing prices. The data is now cleaned, preprocessed, and includes new engineered features, making it suitable for predictive modeling.
